**Задача 1**. Дано слово из латинских букв. Напишите скрипт, который выводит на экран букву из середины слова (если число букв нечетное). Если букв четное число, то на экран выводятся две буквы из середины.  

Пример: для 'test' должно выводится 'es', для 'testing' - 't'

```python
# Запрашиваем у пользователя информации и записываем её в переменную
x = input('Введите слово: ')

# назначаем переменную в которой храним количество букв
in_number = len(x)
# количество букв делим пополам
half_words = in_number / 2
# в случаее если число нечетное, выводим две буквы
half_words_two = in_number / 2 + 1

# ставим условие, если при делении слова будет 0 букв
if in_number % 2 == 0:
    # выводим две буквы из слова
    print(x[int(half_words) - 1: int(half_words_two)])
    # в любых других случаях одну букву
else:
    print(x[int(half_words)])
```

**Задача 2**. Мы делаем MVP dating-сервиса, и у нас есть список парней и девушек (их число может варьироваться):

```python
boys = ['Peter', 'Alex', 'John', 'Arthur', 'Richard']
girls = ['Kate', 'Liza', 'Kira', 'Emma', 'Trisha']

# сортировка и назначение в новую переменную
boys_sort = sorted(boys)
girls_sort = sorted(girls)
```

```python
# Проверка сортировки
print(boys_sort)
print(girls_sort)
```

```python
# назначение переменной с зип объектом состоящим из двух отсортированных списков
boys_and_girls = zip(boys_sort, girls_sort)
```

```python
# проверка зип объекта
print(list(boys_and_girls))
```

```python
print('Идеальные пары:')

# создаем цикл, в котором происходит выборка по первому и второму элементу в зип объекте
for b, g in zip(boys_sort, girls_sort):
    print(f'{b} и {g}')
```

```python
if len(boys) != len(girls):
    print('Есть люди которые могут остаться без пары')
else:
    print('Все люди в списке имеют пару')
```

Выдвигаем гипотезу: лучшие рекомендации мы получим, если просто отсортируем имена по алфавиту и познакомим людей с одинаковыми индексами после сортировки! "Познакомить" пары нам поможет функция zip, а в цикле распакуем zip-объект и выведем информацию в виде:

Идеальные пары:  
Alex и Emma  
Arthur и Kate  
John и Kira  
Peter и Liza  
Richard и Trisha  

**Внимание! Если количество людей в списках будет не совпадать, то мы никого знакомить не будет и выведем пользователю предупреждение, что кто-то может остаться без пары!**

**Задача 4.** Дан поток логов по количеству просмотренных страниц для каждого пользователя. Список отсортирован по ID пользователя. Вам необходимо написать алгоритм, который считает среднее значение просмотров на пользователя. 
Т. е. надо посчитать отношение суммы всех просмотров к количеству уникальных пользователей.

```python
stream = [
    '2018-01-01,user1,3',
    '2018-01-07,user1,4',
    '2018-03-29,user1,1',
    '2018-04-04,user1,13',
    '2018-01-05,user2,7',
    '2018-06-14,user3,4',
    '2018-07-02,user3,10',
    '2018-03-21,user4,19',
    '2018-03-22,user4,4',
    '2018-04-22,user4,8',
    '2018-05-03,user4,9',
    '2018-05-11,user4,11',
]
```

```python
# назначаем уникального пользователя в переменную
id = input('Введите ID пользователя (например, user1): ')

# назначаем переменную куда будет прозводиться суммирование со второго индекса[2] списка с данными
sum_views = []

# строим цикл
for i in stream:
    # внутри цикла делаем в каждой строке разбивку по запятой
    i = i.split(',')
    # ставим условие, где происходит выборка по юзеру (условие в данном месте сократит
    # подсчет суммы просмотров пользователей)
    if i[1] == id:
        # продолжаем внутри цикла складывает все значения по индексу [2] в "виртуальную" переменную
        x = int(i[2])
        # "виртуальную" переменную с результатом добавляем в список
        sum_views.append(x)
        # производим остальные выяесления для среднего значения
        user_mean = sum(sum_views) / len(sum_views)
        
# выводим среднее значение по условию
print(f'Среднее число просмотров выбранным пользователем: {user_mean}')

# Проверка переменной
# print(sum_views)
# print(type(sum_views))
```

**Задача 5**. Дана статистика рекламных кампаний по дням. Напишите алгоритм, который по паре дата-кампания ищет значение численного столбца. 
Т. е. для даты '2018-01-01' и 'google' нужно получить число 25. 
Считайте, что все комбинации дата-кампания уникальны.

```python
stats = [
    ['2018-01-01', 'google', 25],
    ['2018-01-01', 'yandex', 65],
    ['2018-01-01', 'market', 89],
    ['2018-01-02', 'google', 574],
    ['2018-01-02', 'yandex', 249],
    ['2018-01-02', 'market', 994],
    ['2018-01-03', 'google', 1843],
    ['2018-01-03', 'yandex', 1327],
    ['2018-01-03', 'market', 1764],
]

# Запрашиваем у пользователя дату
date = input('Введите дату (гггг-мм-дд): ')
# запрашиваем у пользователя название компании
company = input('Введите название компании: ')
# на всякий случай вводимое в компании приводим к нижнимо риестру
company_format = company.lower()

# делаем цикл по списку
for i in stats:
    # где соблюдается условие, там где дата из списка совпадает с вводимой датой пользователя
    # и с вводимой компанией
    if (date == i[0]) and (company_format == i[1]):
        # выводим результат
        print(i[2])
```

**Задача 6**. Дан список вида:

```python
data = [
    [13, 25, 23, 34, 33],
    [45, 32, 44, 47, 55],
    [12, 33, 23, 95, 88],
    [13, 53, 34, 35, 90],
    [13, 53, 34, 35, 90]
]
# print(data[0])
# print(type(data))
```

Напишите код, который будет вычислять сумму элементов на диагонали. Т. е. 13+32+23+35.  
Список может быть любой длины, но всегда является "квадратным" (количество элементов во вложенных списках равно их количеству).

```python
# Назначаем переменную , в которой отмечены начальные позиции пересчета
# в данный момент индекс первого элемента списка [0][0]
i = 0
# назначаем переменную в которой будет происходить сложение полученных результатов
sum_x = 0

# создаем цикл
for el in data:
    # где, в переменную с каждым новым элементом происходит возрастание, изначально в элементе 0
    # в зависимости от i
    sum_x += el[i]
    # print(el[i])
    # возрастание значения i на 1
    i += 1
    # по окончанию цикла выводим собранную сумму в переменной
print(sum_x)


# обнулим переменную
del sum_x
```

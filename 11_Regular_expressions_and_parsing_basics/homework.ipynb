# Задание 1.
Напишите функцию, которая принимает на вход строку и проверяет является ли она валидным транспортным номером.  
Если да, то функция должна возвращать отдельно номер и регион.

```python
# Решение
import re
number = 'в990ск 96RUS, а555ук 78RUS, 78RUS р666АВ, в236ск 78R, в265ук 65RUS'
number_two = 'в990ск 96RUS'
```

```python
def valid_number(List_):
    """Проверка на валидность номера, с выводом дробленым на две части - номер и регион"""
    
    number_true = str(re.findall(r'([а-я]\d{3}[а-я]{2}\s\d{2}[A-Z]{3})', List_))
    result_first = re.findall(r'([а-я]{1}\d{3}[а-я]{2})', number_true)
    result_second = re.findall(r'(\d{2}[A-Z]{3})', number_true)
    print(result_first) 
    print(result_second)
```

```python
valid_number(number)
```

```python
valid_number(number_two)
```

# Задание 3.


Напишите функцию, которая будет удалять все последовательные повторы слов из заданной строки при помощи регулярных выражений

```python
text = 'hjhjjjjhh rerttyt?///...******##########**********AAAAA66666666********'
```

```python
def no_matches(list_):
    """Удаление совпадающих элементов в списке"""
    
    regex = r'(.)\1+'
    text_format = re.sub(regex, r'\1', list_, 0, re.MULTILINE)
    print(text_format)
```

```python
no_matches(text)
```

Напишите функцию, которая будет проверять номер сотового телефона на валидность, если он валиден, то переводить его в формат:
+7(xxx)-xxx-xx-xx


Постарайтесь предусмотреть как можно больше адекватных форматов изначального ввода номера. Примеры валидных форматов:


+7 955 555-55-55;
9555555555;
8(955)555-55-55;
+7 955 555 55 55;
7(955) 555-55-55;
+7 955+555+55+55.

```python
number_phone = '+7 955 555-55-55, 95555555555, 8(955)555-55-55, +7 955 555 55 55, 7(955) 555-55-55, +7 955+555+55+55'
```

```python
def valid_number(list_):
    """Проверка на валидность телефоного номера, и выводом его в стандартном формате"""
    
    number_phone_true = str(re.findall(r'\W\d{1} \d{3}\W\d{3}\W\d{2}\W\d{2}', list_))
    number_phone_true1 = str(re.findall(r'\d{1}\d{3}\d{3}\d{2}\d{2}', list_))
    number_phone_true2 = str(re.findall(r'\d{1}\(\d{3}\)\d{3}-\d{2}-\d{2}', list_))
    number_phone_true3 = str(re.findall(r'\d{1}\(\d{3}\) \d{3}-\d{2}-\d{2}', list_))

    standart_number = str(re.sub(r'\W(\d{1}) (\d{3})\W(\d{3})\W(\d{2})\W(\d{2})', r'+\1(\2)-\3-\4-\5', number_phone_true))
    standart_number1 = str(re.sub(r'(\d{1})(\d{3})(\d{3})(\d{2})(\d{2})', r'+\1(\2)-\3-\4-\5', number_phone_true1))
    standart_number2 = str(re.sub(r'(\d{1})\W(\d{3})\W(\d{3})-(\d{2})-(\d{2})', r'+\1(\2)-\3-\4-\5', number_phone_true2))
    standart_number3 = str(re.sub(r'(\d{1})\((\d{3})\) (\d{3})-(\d{2})-(\d{2})', r'+\1(\2)-\3-\4-\5', number_phone_true3))
    
    print(standart_number)
    print(standart_number1)
    print(standart_number2)
    print(standart_number3)
```

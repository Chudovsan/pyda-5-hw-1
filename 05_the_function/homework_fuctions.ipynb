```python
# p – people – команда, которая спросит номер документа и выведет имя человека, которому он принадлежит

# Назначаем функцию
def p():
    
    # при вызове функции запрашиваем у пользователя номер документа 
    # и помещаем его в переменную функции document_number
    document_number = input("Введите номер документа: ")

    # задаем цикл с перебором строк  
    for doc in documents:
        # так же запускаем внутри ещё цикл на перебор строк в doc
        for number in doc:
            # ставим условие где,  ключ number равен переменной функции document_number
            if document_number == doc[number]:
                # возвращаем значение ключа name
                return doc['name']
    # в случае отсутствия указанного номера в словаре документов
    # выводим комментарий
    print(f'Документа {document_number} нет')

# вызов функции       
#p()
```

```python
#s – shelf – команда, которая спросит номер документа и выведет номер полки, на которой он находится
#Правильно обработайте ситуацию, когда пользователь будет вводить несуществующий документ.

# Назначаем функцию
def s():
    # запрашиваем у пользователя номер документа
    document_number = input("Введите номер документа: ")
 
    # задаем цикл с перебором строк в словаре полок
    for k in directories:
        
        # ставим условие где, просмотреть document_number на наличие в словаре полок
        if document_number in directories[k]:
            # и отобразить полку с совпадением
            return k
    # в случае ввода документа котоый не существует
    print(f'Документ {document_number} не существует')
    
# вызов функции
#s()
```

```python
#l – list – команда, которая выведет список всех документов в формате
#passport "2207 876234" "Василий Гупкин"        

# Назначаем функцию 
def l():
    
    # запускаем цикл 
    for x in documents:
        # в переменные внутри цикла записываются значения по ключам
        type = x['type']
        number = x['number']
        name = x['name']
        # шаблон отображения
        print(f'{type}, {number}, {name}')
        
#  вызов функции с выводом значений по установленому шаблону
#l()
```

```python
# as – add shelf – команда, которая спросит номер новой полки и добавит ее в перечень
# Назначаем функцию
def as_():
    # при вызове функции, будет происходить запрос пользователю
    new_shelf = input('Введите номер новой полки: ')
    
    # проверка на существование полки с таким номером
    # ставим условие если new_shelf уже существует в ключах словаря полок
    if new_shelf in directories.keys():
        # отображение полок
        #print(directories)
        # возвращает номер полки с комментарием
        return(f'Полка № {new_shelf} уже существует.')
    # в любом другом случае
    else:
        # создает новыу запись в словаре с указаным ключом
        directories[new_shelf] = []
        # просмотр изменений в полках
        #print(directories)
        # возвращаем указанный комментарий с номером полки
        return(f'Полка с номером {new_shelf} создана.')
    
#as_()
```

```python
#d – delete – команда, которая спросит номер документа и удалит его из каталога и из перечня полок. 
#Предусмотрите сценарий, когда пользователь вводит несуществующий документ;

# Назаначаем функцию
def d():
    # при вызове функции делаем запрос к пользователю и назначаем в переменную
    doc_number = input('Введиет номер документа, который следует удалить: ')


    # удаление из словаря документов
    # запускаем цикл с перебором строк
    for item in documents:
        # если по ключу number в словаре документов есть совпадения
        if item['number'] == doc_number:
            # строка удаляется
            documents.remove(item)


    # удаление из словаря с полоками
    # запускаем цикл с разделением на несколько временных переменных данных в словаре полок
    for key, value in directories.items():
        # так же внутри цикла запускаем цикл на перебор номеров документов на полках
        for i in value:
            # если номер документа совпадает с введеным значением
            if i == doc_number:
                # удаляем номер документа с полки
                value.remove(i)
                #print(documents)
                #print(directories)
                return 'Документ удален из директории документов и директории полок'
            
    #print(documents)
    #print(directories)
    # в случае если номер не совпадает, возвращаем коментарий
    return 'Такого документа нет.'
#d()
```

```python
#m – move – команда, которая спросит номер документа и целевую полку и переместит его с текущей полки на целевую. 
#Корректно обработайте кейсы, когда пользователь пытается переместить несуществующий документ 
#или переместить документ на несуществующую полку

#Назначаем функцию
def m():
    # после вызова функции идет запрос у пользователя
    doc_number = input('Введите номер документа который следует перенести на другую полку: ')
    directories_number = input('Введите номер полки на которую следует перенести выбранный документ: ')
    
    # при условии если directories_number нет в ключах словаря полок
    if directories_number not in directories.keys():
        # возвращаем комментарий
        return(f'Полки {directories_number} не существует')
    # в других случаях 
    else:
        # запускаем цикл по строкам словаря документов
        for item in documents:
            # если строка с ключом number равен с вводимым значением пользователя
            if item['number'] == doc_number:
                # запускаем цикл по значениям словаря полок
                for value in directories.values():
                    # где просматриваем указанный номер в значениях словаря полок 
                    if doc_number in value:
                        # и удаляем
                        value.remove(doc_number)
                # добавляем строку с указанным номером в указанный ключ полок 
                directories[directories_number].append(doc_number)
                #print(documents)
                #print(directories)
                # возвращаем комментарий с указанной полкой
                return(f'Документ перенесен на полку № {directories_number}')
            
        #print(documents)
        #print(directories)
        # возвращаем комментарий если указанного документа нет
        return(f'Номера документа {doc_number} не существует.')
#m()
```

```python
#a – add – команда, которая добавит новый документ в каталог и в перечень полок, 
#спросив его номер, тип, имя владельца и номер полки, на котором он будет храниться.

# назначим функцию
def a():
    # после вызова функции идет запрос к пользователю
    doc_type = input('Введите тип документа: ')
    doc_number = input('Введите номер документа: ')
    doc_holder_name = input('Введите имя владельца документа: ')
    directories_number = input('Введите номер полки для данного документа: ')

    # ставим условие, если directories_number не совпадает с каким нибудь ключом словаря полок 
    if directories_number not in directories.keys():
        #print(documents)
        #print(directories)
        # возвращаем комментарий 
        return(f'Полки {directories_number} не существует')
    # в других случаях 
    else:
        # добавление нового документа в каталог документов
        documents.append({
            'type': doc_type,
            'number': doc_number,
            'name': doc_holder_name
        })

        # добавление номера документа на полку
        directories[directories_number].append(doc_number)
        #print(documents)
        #print(directories)
        return(f'Документ добавлен на полку {directories_number}')

#a()
```

```python
# назначаем функцию в которой перечисляем возможные варианты ввода
def menu(input_):
    # при совпадении вызываем функции и выводим результат их работы
    if input_ == 'p':
        print(p())
    elif input_ == 'l':
        l()
    elif input_ == 's':
        print(s())
    elif input_ == 'a':
        print(a())
    elif input_ == 'd':
        print(d())
    elif input_ == 'm':
        print(m())
    elif input_ == 'as':
        print(as_())
        # при любом другом вводе
    else:
        print('Введена не верная команда.')
```

```python
# выводим инструкцию для пользователя

print('''

p – вывод имя человека по номеру документа

l – вывод списока всех документов

s – вывод номера полки на которой находится документ

a – добавить новый документ в каталог и в перечень полок, спросив его номер, 
тип, имя владельца и номер полки, на котором он будет храниться

d – удалить документ из каталога и из перечня полок

m – переместить документ с полки на полку

as – добавить новую полку

q - выход

''')

# запускаем бесконечный цикл
while True:
    # запрос пользователю
    input_ = input ('\nВведите команду: ').lower()
    # если q программа останавливается
    if input_ == 'q':
        break
    # остальное исполнение функции menu
    else:
        menu(input_)
```

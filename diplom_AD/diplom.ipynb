### ACTUALDATE, TRANSDATE - object in datetime64

###### ACTUALDATE

```python
df1['ACTUALDATE']
```

```python
# замена г., на пробел
df1['ACTUALDATE'] = df1['ACTUALDATE'].str.replace('г.,', ' ')
```

```python
#удаляем пробелы
df1['ACTUALDATE'] = df1['ACTUALDATE'].str.strip()
```

```python
df1['ACTUALDATE']
```

```python
# делим на части
df1['ACTUALDATE'] = df1['ACTUALDATE'].str.split()
```

```python
df1['ACTUALDATE'].head(2)
```

```python
# разделим столбец на несколько нужных списков, для обработки месяцев
```

```python
day = []
mouth = []
years = []
time = []

for a, b, c, e in df1['ACTUALDATE']:
    #print(a, b, c, e)
    a = int(a)
    b = str(b)
    c = int(c)
    e = str(e)
    day.append(a)
    mouth.append(b)
    years.append(c)
    time.append(e)
```

```python
# соеденим в один 
data_result = []

for i in zip(day, mouth, years, time):
    #print(i)
    data_result.append(i)
```

```python
# запишим в дата фрейм для удобства
df2 = pd.DataFrame(data_result, columns=['Day', 'Mouth', 'Years', 'Time'])
```

```python
df2.head(2)
```

```python
# Просмотрим уникальные значения в месяце
df2.Mouth.unique()
```

```python
# Перепишем
df2['Mouth'] = df2['Mouth'].str.replace('янв.', 'Jan')
df2['Mouth'] = df2['Mouth'].str.replace('февр.', 'Feb')
df2['Mouth'] = df2['Mouth'].str.replace('марта', 'Mar')
df2['Mouth'] = df2['Mouth'].str.replace('апр.', 'Apr')
df2['Mouth'] = df2['Mouth'].str.replace('мая', 'May')
df2['Mouth'] = df2['Mouth'].str.replace('июня', 'June')
df2['Mouth'] = df2['Mouth'].str.replace('июля', 'July')
df2['Mouth'] = df2['Mouth'].str.replace('авг.', 'Aug')
df2['Mouth'] = df2['Mouth'].str.replace('сент.', 'Sept')
df2['Mouth'] = df2['Mouth'].str.replace('окт.', 'Oct')
df2['Mouth'] = df2['Mouth'].str.replace('нояб.', 'Nov')
df2['Mouth'] = df2['Mouth'].str.replace('дек.', 'Dec')
```

```python
# проверим
df2.Mouth.unique()
```

```python
# Перейдем ко времени
df2['Time']
```
